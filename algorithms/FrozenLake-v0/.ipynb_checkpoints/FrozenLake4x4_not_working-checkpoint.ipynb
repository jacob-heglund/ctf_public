```python
## "deep" Q-network 
class myDQN(nn.Module):
    def __init__(self, num_states, num_actions):
        super(myDQN, self).__init__()
        self.fc1 = nn.Linear(num_states, num_actions)
        self.relu = nn.ReLU(inplace=True)
        
    def forward(self, x):
        out = self.fc1(x)
        out = self.relu(out)
        return out
```

```python
## convert state to a vector to be fed into the Q-function approximator
def int_to_one_hot(state, state_space_size):
    obs = torch.zeros([1, state_space_size])
    obs[0, state] = 1
    return obs

def one_hot_to_int(state_one_hot):
    state_int = np.argmax(state_one_hot)
    return state_int

## returns a decaying chance of taking random actions
def epsilon_by_frame(frame_idx):
    epsilon_curr = epsilon_final + (epsilon_start - epsilon_final) * math.exp(-1. * frame_idx / epsilon_decay)
    return epsilon_curr
```

```python
## take an epsilon greedy action 
def gen_action(state):
    #TODO add to GPU
    if np.random.rand(1) < epsilon:
        q_values = online_model(state)
        max_q_value, _= torch.max(q_values, 1)        
        action = env.action_space.sample()
    
    else:
        q_values = online_model(state)
        max_q_value, action = torch.max(q_values, 1)
        action = action.item()
        
    return action, max_q_value
```

```python
## play an episode until done
def play_episode(epsilon):
    episode_length = 0.
    episode_reward = 0.
    episode_loss = 0.
    done = 0
    state = env.reset()
    
    while (done == 0):
        episode_length += 1
        state = int_to_one_hot(state, num_states)
        action, q_value = gen_action(state)
        next_state, reward, done, _ = env.step(action)
        
        if (done == 1) and (reward == 0):
            reward = -1.0
            
        # TODO append to replay buffer here

        #TODO add torch.no_grad() here
        # Q-learning
        with torch.no_grad():
            next_state = int_to_one_hot(next_state, num_states)
            next_action, next_q_value = gen_action(next_state)
            target = reward + torch.mul(next_q_value, gamma)
        
        # training
        loss = criterion(q_value, target)
        online_model.zero_grad()
        loss.backward()
        optimizer.step()

        state = one_hot_to_int(next_state)
        episode_reward += reward
        episode_loss += loss.item()
        
        if done == True:
            if reward > 0:
                epsilon = 1./((episode/50)+10)
            break
    episode_loss = episode_loss / episode_length
    
    return episode_length, episode_reward, episode_loss, epsilon
```

```python
## training
#TODO add decaying epsilon
#TODO add experience replay -> sample batches of actions
epsilon = 0.05
gamma = 0.99
LR = 0.01
num_episodes = 1000

num_states = env.observation_space.n
num_actions = env.action_space.n

online_model = myDQN(num_states, num_actions)
criterion = nn.MSELoss()
optimizer = optim.Adam(online_model.parameters(), lr=LR)

# save training data
length_list = []
reward_list = []
loss_list = []
epsilon_list = []

for episode in trange(num_episodes):
    episode_length, episode_reward, episode_loss, epsilon = play_episode(epsilon)
    
    length_list.append(episode_length)
    reward_list.append(episode_reward)
    loss_list.append(episode_loss)
    epsilon_list.append(epsilon)
    
print('\nPercent Successful Episodes: {}'.format(np.sum(np.array(reward_list)>0.0)/num_episodes))
```
